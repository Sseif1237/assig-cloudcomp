```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('D:/cloudcomputing/books.csv')
df.head()


```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values
df.dropna(inplace=True)

# Drop duplicate rows
df.drop_duplicates(inplace=True)
# Focus analysis on the Harry Potter series by filtering the dataset
harry_potter_df = df[df['original_title'].str.contains('Harry Potter', case=False)]

# Convert sales column to numeric 
harry_potter_df['ratings_count'] = pd.to_numeric(harry_potter_df['ratings_count'], errors='coerce')
# Find the most selling books within the Harry Potter series
most_rating_books = harry_potter_df.sort_values(by='ratings_count', ascending=False).head(5)
# Calculate the average rating of the Harry Potter books
average_rating = harry_potter_df['average_rating'].mean()
# Display results
print("Most rating Books in the Harry Potter Series:")
print(most_rating_books)
print("\nAverage Rating of the Harry Potter Books:", average_rating)
# Save the cleaned data to new csv file
harry_potter_df.to_csv("harry_potter_books.csv", index=False)






```
